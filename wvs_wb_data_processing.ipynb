#### function ```CreateSelData```

````python
def CreateSelData(df, wave):

    info_list = 'info_list_' + wave
    trust_list = 'trust_list_' + wave
    cov_list = 'cov_list_' + wave
    
    info_dict = globals()['info_dict_' + wave]
    trust_dict = globals()['trust_dict_' + wave]
    cov_dict = globals()['cov_dict_' + wave]
    
    locals()[info_list] = list(info_dict.keys())
    locals()[trust_list] = list(trust_dict.keys())
    locals()[cov_list] = list(cov_dict.keys())
    
    var_list = locals()[info_list] + locals()[trust_list] + locals()[cov_list]  
    var_dict = {**info_dict, **trust_dict, **cov_dict}
    
    df_sel = df[var_list]
    df_sel_final = df_sel.rename(columns=var_dict)
    
    return df_sel_final
````

### check duplicated responses ```CheckDupID```

````python
def CheckDupID(df):
    grouped = df.groupby(['country_iso','year', 'interview_ID']).size().reset_index(name='count')
    
    total_responses = df.groupby(['country_iso','year']).size().reset_index(name='total_count')
    
    merged = pd.merge(grouped, total_responses, on=['country_iso','year'])
    
    duplicated_responses = merged[merged['count'] > 1].groupby(['country_iso','year'])['count'].sum().reset_index(name='dup_count')
    
    result = pd.merge(total_responses, duplicated_responses, on=['country_iso', 'year'], how='left').fillna(0)
    
    result['dup_percent'] = round(result['dup_count']/result['total_count']*100,2)
    
    result_dup = result.query('dup_count > 0')
    
    return result_dup
````

### check frequencies of variables ```CheckFreq```

````python
def CheckFreq(df):
    frequency_tables = {}
    
    for column in df.columns:
        frequency_tables[column] = df[column].value_counts()
    
    for column, frequencies in frequency_tables.items():
        print(f"Frequency table for {column}:")
        print(frequencies)
        print()
````

````python
n_missing_list = []
n_non_missing_list = []
percent_missing_list = []

for column in df_master.columns:
    n_missing = df_master[column].isna().sum()
    n_non_missing = df_master.shape[0] - n_missing
    percent_missing = n_missing/df_master.shape[0]
    
    n_missing_list.append(n_missing)
    n_non_missing_list.append(n_non_missing)
    percent_missing_list.append(percent_missing)
    
    col_list = list(df_master.columns)
    
df_master_missing = pd.DataFrame({
    'variable': df_master.columns,
    'n_missing': n_missing_list, 
    'n_non_missing': n_non_missing_list,
    'percent_missing': percent_missing_list
})
````

### recode multi-categorical variables ```employment```,```marital```,```education``` into binary
### recode ```gender``` to be 0 vs. 1

````python
df_master['employment_binary'] = np.where(
    df_master['employment']==7,0,
    np.where(df_master['employment'].isna(), np.nan, 1)) #1 = employed, #0 = unemployed
````

````python
df_master['marital_binary'] = np.where(
    df_master['marital']==1,1,
    np.where(df_master['marital'].isna(), np.nan, 0)) #1 = married, #0 = not married
````

````python
if (df_master['wave'] == 7).any():  # Check if all values in 'wave' column are 7
    df_master['hs_binary'] = np.where(
        df_master['education'] <= 2, 0,
        np.where(df_master['education'].isna(), np.nan, 1))
else:
    df_master['hs_binary'] = np.where(
        df_master['education'] <= 6, 0,
        np.where(df_master['education'].isna(), np.nan, 1)) #1 = high school graduate, #0 = less than high school
````

````python
df_master['gender_recoded'] = np.where(
        df_master['gender'] == 2, 0,
        np.where(df_master['gender'].isna(), np.nan, 1)) #1 = male #0 = female
````

### cluster(country)-centering & grand-mean-centering relevant continous IVs

````python
cat_iv_list = ['gender_recoded', #1 = male, #0 = female
               'employment_binary',
               'marital_binary',
               'hs_binary']
````

````python
con_iv_list = ['settlement_size', #essentially ordinal
               'sub_class', #1 = upper, 2 = uppper-mid, 3 = lower-mid, 4 = working 5 = lower ##essentially ordinal
               'sub_class_recoded',
               'ideology', #1 = left, 7 = right
               'age', 'income',
               'pop_density', 'total_pop',
               'gdp_2015usd',
               'poverty_ratio', 
               'gini', 
               'total_pop',
               'unemployment'] 
````

````python
for var in con_iv_list:
    
    var_cmc = var + '_cmc'
    df_master[var_cmc] = df_master.groupby('country_iso')[var].transform(lambda x: x - x.mean())
````

````python
for var in con_iv_list:
    var_gmc = var + '_gmc'
    df_master[var_gmc] = df_master[var].transform(lambda x: x - x.mean())
````
